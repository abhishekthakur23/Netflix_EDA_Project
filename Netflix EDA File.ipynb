# VISUALIZATIONS

```python
df.type.value_counts()
```

```python
sns.countplot(x="type", data = df)
plt.title("Count Vs Type of Shows")
```

```python
sns.countplot(y="listed_in", order = df["listed_in"].value_counts().index[0:15], data = df)
plt.title("Top 20 Genres on Netflix")
```

```python
# Aggregate the data by date to count the number of shows/movies added each day
time_series_data = df.groupby('date_added').size().reset_index(name='count')

# Set the date as the index
time_series_data.set_index('date_added', inplace=True)

# Display the first few rows of the aggregated data
print(time_series_data.head())
```

```python
plt.plot(time_series_data.index, time_series_data['count'], marker='o')
plt.title('Number of Shows/Movies Added to Netflix Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Shows/Movies Added')
plt.show()
```

```python
sns.countplot(data = df, y="listed_in", order = df["listed_in"].value_counts().index[0:20])
```

```python
plt.figure(figsize = (10,7))
sns.countplot(data=df, x="release_year", order = df["release_year"].value_counts().index[0:20])
plt.title("Distribution of content across release years")
```

```python
sns.countplot(data = df, y="country", order = df["country"].value_counts().index[0:10])
plt.title("Distribution of content across different countries and regions")
```
